```python
import cv2
import numpy as np
import matplotlib.pyplot as mp
```

```python
image1=cv2.imread("dog.jpg")
image2=cv2.imread("dog1.jpg")
image1=cv2.cvtColor(image1,cv2.COLOR_BGR2RGB)
image2=cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)
```

```python
mp.imshow(image1)
```

```python
mp.imshow(image2)
```

```python
image1_array=image1.copy()
image2_array=image2.copy()
```

```python
image1_array_cropped=image1_array[50:150,50:150]
image1_array_cropped.shape
```

```python
mp.imshow(image1_array_cropped)
cv2.imwrite("image1.jpg",image1_array_cropped)
```

```python
image2_array_cropped=image2_array[50:150,50:150]
image2_array_cropped.shape
```

```python
mp.imshow(image2_array_cropped)
cv2.imwrite("image2.jpg",image2_array_cropped)
```

```python
image1[50:150,50:150]=image2_array_cropped
```
